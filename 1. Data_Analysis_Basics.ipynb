## b) Statistical Insights:
With this next step, we will do a quick **Statistical Summary** of each of our columns to find out about the distribution of data within them.

```python
# Using the .describe method to get a Statistical Summary:
df.describe
```

Note that the Rows & Columns that do not contain numbers are skipped. However, to get a summary of all columns, the **"all"** arguments can be added to the **.descibe** method.

```python
# Displaying the Full Summary:
df.describe(include="all")
```

#### By using the **.describe** method,we can discover any mathematical anomalies such as big deviations and extreme outliers in our data.
#### It returns:

1. **Count** - the number of terms in a column.
2. **Mean** - the average column value.
3. **Std** - Standard Deviation of Columns
4. Max. & Min. values.
5. The boundaries of each Quartile.

#### **Unique, Top & Frequency** represent the evaluation of Columns containing Object-Type.
6. **Unique** refers to the number of distinct objects in a column.
7. **Top** indicates the object that appears most frequently.
8. **Freq** is the number of times that the top object is appearing in a given column.
